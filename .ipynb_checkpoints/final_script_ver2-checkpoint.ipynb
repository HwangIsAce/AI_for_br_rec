```python
def cal(n1,n2):
    cnt = 0
    column1 = []
    column_list = list(df.columns[1:])
    total = len(column_list)

    for idx in range(1,len(column_list)):
        if ((check.iloc[n1,idx]==0 and check.iloc[n2,idx]>0) or (check.iloc[n1,idx]>0 and check.iloc[n2,idx]==0)):
        #둘 중 하나만 0인 경우
            column1.append(column_list[idx])
    
    #위의 경우를 제외한 나머지 경우만 표본으로 잡고, 계산
    l = list(set(column_list)-set(column1))
    for idx in l:
        if check.iloc[n1][idx]>=35 and check.iloc[n2][idx]>=35: #예측,정답 함량이 모두 35이상인 경우
            cnt+=1
        elif check.iloc[n1][idx]>=10 and check.iloc[n1][idx]<35: #예측이 10과 35사이
            if check.iloc[n2][idx]>=10 and check.iloc[n2][idx]<35: #정답도 10과 35사이
                cnt+=1
        elif check.iloc[n1][idx]<10 and check.iloc[n1][idx]>0: #예측이 0과 10 사이
            if check.iloc[n2][idx]<10 and check.iloc[n2][idx]>0: #정답이 0과 10 사이
                cnt+=1
        elif check.iloc[n1][idx]==0 and check.iloc[n2][idx]==0: #예측과 정답이 0일 때
            cnt+=1
        
    result = cnt/total
    return result
```

```python
# 가장 점수가 높은 레시피를 선택함
score = (cal(0,3)+cal(1,4)+cal(2,5)) /3

print(f"score:{score}")
```
